# Creating Scattar Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure matplotlib plots show in the notebook
%matplotlib inline

import pandas as pd
```

```python
data = pd.read_csv('/kaggle/input/data-for-datavis/candy.csv')
data.head()
```

```python
x = list(data['winpercent'])
y = list(data['pricepercent'])
z = list(data['sugarpercent'])

plt.scatter(x,y)
plt.scatter(x,z)

plt.show()
```

# Customizing the scatter plot

```python
x = list(data['winpercent'])
y = list(data['pricepercent'])
z = list(data['sugarpercent'])

plt.scatter(x,y)
plt.scatter(x,z)

# Adding legends

plt.legend(['y', 'z'], loc = 'best')

# adding title, label

plt.title('Scatter plot Demo')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

# Color, Edge Color, Size, Opacity, Marker Shape

```python
# alpha = opacity
# s = size
# marker = shape
plt.scatter(x,y, color = 'green', edgecolor = 'yellow', alpha = 0.5, s = 300, marker = 'o')

plt.scatter(x,z, color = 'red', edgecolor = 'black', alpha = 0.3, s = 200, marker = 'o')

# Adding legends

plt.legend(['y', 'z'], loc = 'best')
plt.show()
```
