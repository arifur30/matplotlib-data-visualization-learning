# Creating a Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creating a dataset
data = {
    'Apple': 50,
    'Banana': 35,
    'Orange': 45,
    'Mango': 25,
    'Grape': 40
}

names = list(data.keys())
values = list(data.values())

plt.figure(figsize = (12,5))

# creating the bar

plt.bar(names, values)
plt.show()
```

```python
data = {
'Fruit': ['Apple', 'Banana', 'Orange', 'Mango', 'Grape'],
'Tons Produced': [50, 35, 45, 25, 40]
}


names = list(data['Fruit'])
values = list(data['Tons Produced'])

plt.figure(figsize = (12,5))

# creating the bar

plt.bar(names, values)
plt.show()
```

# Customizing the Bar chart

```python
plt.bar(names, values)

# Labeling

plt.title("Fruits Distribution")
plt.xlabel("Fruits name")
plt.ylabel("Total Produced")
plt.show()
```

```python
# Color

plt.bar(names, values, color = 'orange')
plt.show()
```

**Horizontal Bar**

```python
plt.barh(names, values, color = 'orange')  #just use 'barh'
plt.show()
```
