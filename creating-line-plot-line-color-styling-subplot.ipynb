# Plot axis labeling


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,10,0.1)
y = 2*x + 5

plt.plot(x,y)

# plot axis labeling
plt.title('First plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show
```

# Line styling


```python
plt.plot(x,y, linewidth = 2.0, linestyle = ':')
```

```python
plt.figure(figsize=(12,6))

# Dashed line
plt.plot(x, y, linestyle='--', label='Dashed Line')


plt.legend()
plt.show()
```

```python
# Dotted line
plt.plot(x, y, linestyle=':', label='Dotted Line')

plt.legend()
plt.show()
```

```python

# Dash-dot line
plt.plot(x, y, linestyle='-.', label='Dash-Dot Line')
plt.legend()
plt.show()
```

# Line Color

# Grid & Legend


```python
plt.plot(x, y, color = 'green')

plt.legend(['Green Line'], loc = 'best')  # if the line is not labeled yet
plt.grid(True)
```

# Creating two or more plots in a single fig

```python
plt.figure(figsize=(12,6))


y1 = 2*x + 3
y2 = 3*x + 5

#sub plotting 

plt.subplot(2, 2, 1)  # (rows, columns, plot number)
plt.plot(x, y1 , label = 'Line1', color = 'r')
plt.legend()

plt.subplot(2,2, 2)
plt.plot(x, y2 , label = 'Line2', color = 'g')
plt.legend()

plt.show()
```
